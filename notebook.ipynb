```python
import matplotlib.pyplot as plt
import numpy as np
from math import gcd

# Create a grid of values for a and b
a_values = np.arange(1, 15)
b_values = np.arange(1, 16)

# Create a meshgrid for plotting
A, B = np.meshgrid(b_values, a_values)  # Swap a_values and b_values

# Initialize a matrix to store the results
results = np.zeros_like(A, dtype=bool)
answer = 0
# Check the condition for each pair (a, b)
for i in range(len(a_values)):
    for j in range(len(b_values)):
        a = a_values[i]
        b = b_values[j]
        if b % gcd(a, 15) == 0:
            answer += 1
            results[i, j] = True  # Swap indices

# Plot the results
plt.figure(figsize=(8, 8))
for i in range(len(a_values)):
    for j in range(len(b_values)):
        if results[i, j]:
            plt.plot(b_values[j], a_values[i], 'x', color='red')  # Swap a_values and b_values

plt.xlabel('b')
plt.ylabel('a')
plt.title('Values of a and b for which ax â‰¡ b (mod 15) has a solution')
plt.xticks(b_values)
plt.yticks(a_values)
plt.grid(True)
plt.show()
print(answer)
# Compute the number of True values in each line (row)
true_counts_per_line = np.sum(results, axis=1)

# Print the counts
for idx, count in enumerate(true_counts_per_line):
    print(f"{count} + ", end="")
```
